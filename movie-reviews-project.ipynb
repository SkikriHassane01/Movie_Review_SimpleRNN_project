```python
import numpy as np 
import pandas as pd 
import tensorflow as tf 
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
from tensorflow.keras.callbacks import EarlyStopping
```

# <center style="color:skyblue; font-size:40px;">Load the imdb dataset</center>

```python
max_features = 10000 # size of vocabulary (number of unique words in your dataset)
(X_train, y_train),(X_test,y_test) = imdb.load_data(num_words = max_features)
```

```python
print(f"Training Data shape : {X_train.shape}, Testing Data Shape: {X_test.shape}")
```

```python
X_train 
```

```python
# mapping of words index to words

word_index = imdb.get_word_index()
reversed_word_index = {value:key for key,value in word_index.items()}

print(reversed_word_index[2])
print(reversed_word_index[4])
```

```python
# based on the documentation of tensorflow this is the method to get the decoded review 
sample_review = X_train[0]

decoded_review = ' '.join([reversed_word_index.get(i - 3, '?') for i in sample_review])
```

```python
print(decoded_review)
```

# <center style="color:skyblue; font-size:40px;">Preprocessing </center>

```python
len_list = []
for i in range(0,25000):
    len_list.append(len(X_train[i]))
```

```python
print(np.min((len_list)))
print(np.max((len_list)))
```

```python
max_len = 500
```

```python
X_train_pad = sequence.pad_sequences(X_train,maxlen=max_len)
X_test_pad = sequence.pad_sequences(X_test,maxlen=max_len)
```

```python
print(len(X_train_pad[0]))
print(len(X_train_pad[1]))
print(len(X_train_pad[2]))
```

# <center style="color:skyblue; font-size:40px;">Train the Simple RNN</center>

```python
model = Sequential([
    Embedding(input_dim=max_features,  # Vocabulary size (max_features)
              output_dim=128),           # Embedding dimension 
    SimpleRNN(128, activation='relu'),  # RNN with 128 units
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# create earlyStopping callback 

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# Train the model with early stopping 

histroy = model.fit(
    X_train_pad,
    y_train,
    epochs=20,
    batch_size=50,
    validation_split=0.2,
    callbacks=[early_stopping]
) 
```

```python
# save the model 
model.save('SimpleRNN_mdb.h5')
```

# prediction

```python
from keras.models import load_model
model = load_model('/kaggle/working/SimpleRNN_mdb.h5')
```

```python
def decode_review(encoded_review):
    return ' '.join([reversed_word_index.get(i -3, '?') for i in encoded_review])

def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = sequence.pad_sequences([encoded_review], maxlen=1000)

    return padded_review
```

```python
def prediction(review):
    preprocessed_input = preprocess_text(review)
    prediction = model.predict(preprocessed_input)
    sentiment = 'Positive' if prediction[0][0] > 0.5 else 'Negative'
    return sentiment, prediction[0][0]    
    
```

```python
# test 

examples = ['This movie was fantastic! the acting was great and teh plot was thrilling.',
           'The movie was very bad']

for _,ex in enumerate(examples):
    sentiment, score = prediction(ex)
    print(f'Review: {ex}')
    print(f'Sentiment: {sentiment}')
    print(f'Prediction Score: {score}')
    
```
